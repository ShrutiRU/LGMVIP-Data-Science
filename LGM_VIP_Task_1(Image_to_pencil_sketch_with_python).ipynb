##**Importing required libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

##**Uploading the required image**

```python
image = cv2.imread("/content/drive/MyDrive/MSD.jpg")
```

##**Displaying original image**

```python
cv2_imshow(image)
```

##**Converting Original Image into Grayscale**

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_gray)
```

##**Transformation of the greyscale image into negative using bitwise_not operator**

```python
image_invert = cv2.bitwise_not(image_gray)
cv2_imshow(image_invert)
```

##**Smoothing the image**

```python
image_smoothing = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(image_smoothing)
```

##**Creating a Pencil Sketch from Original Image**

```python
def dodgeV2(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

##**Displaying the converted pencil sketch**

```python
final_image = dodgeV2(image_gray, image_smoothing)
cv2_imshow(final_image)
```
